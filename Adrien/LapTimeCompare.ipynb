```python
import fastf1 as ff1
from fastf1 import plotting
from matplotlib import pyplot as plt

plotting.setup_mpl()

ff1.Cache.enable_cache('../cache')

race = ff1.get_session(2022, 5, 'R')
laps = race.load_laps()

lec = laps.pick_driver('LEC')
ham = laps.pick_driver('HAM')
ver = laps.pick_driver('VER')
sai = laps.pick_driver('SAI')


fig, ax = plt.subplots(figsize=(15,10))
ax.plot(sai['LapNumber'], sai['LapTime'], color='#DC0000')
ax.plot(ver['LapNumber'], ver['LapTime'], color='#0600EF')
ax.set_title("VER vs SAI")
ax.set_xlabel("Lap Number")
ax.set_ylabel("Lap Time")
plt.show()
```
